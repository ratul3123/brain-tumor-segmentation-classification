```python
import os
import random
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, losses, optimizers, Model, backend as K
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import glob
import gc # Garbage Collector

# --- Configuration ---
DATA_DIR = '/kaggle/input/brisc2025/brisc2025'
# --- Start: Fallback for local testing if Kaggle directory is not found ---
if not os.path.exists(DATA_DIR):
    print("Kaggle directory not found. Creating dummy structure for local testing.")
    DATA_DIR = './brisc2025_dummy'
    # Create dummy directories
    os.makedirs(os.path.join(DATA_DIR, 'segmentation_task/train/images'), exist_ok=True)
    os.makedirs(os.path.join(DATA_DIR, 'segmentation_task/train/masks'), exist_ok=True)
    os.makedirs(os.path.join(DATA_DIR, 'segmentation_task/test/images'), exist_ok=True)
    os.makedirs(os.path.join(DATA_DIR, 'segmentation_task/test/masks'), exist_ok=True)
    classes = ['glioma', 'meningioma', 'no_tumor', 'pituitary']
    for cls in classes:
        os.makedirs(os.path.join(DATA_DIR, 'classification_task/train', cls), exist_ok=True)
        os.makedirs(os.path.join(DATA_DIR, 'classification_task/test', cls), exist_ok=True)
        # Create dummy images
        for i in range(5):
            Image.new('RGB', (100, 100)).save(os.path.join(DATA_DIR, 'classification_task/train', cls, f'dummy_{i}.jpg'))
            Image.new('RGB', (100, 100)).save(os.path.join(DATA_DIR, 'classification_task/test', cls, f'dummy_{i}.jpg'))
    for i in range(20):
        Image.new('RGB', (100, 100)).save(os.path.join(DATA_DIR, 'segmentation_task/train/images', f's_{i}.jpg'))
        Image.new('L', (100, 100)).save(os.path.join(DATA_DIR, 'segmentation_task/train/masks', f's_{i}.png'))
        Image.new('RGB', (100, 100)).save(os.path.join(DATA_DIR, 'segmentation_task/test/images', f's_{i}.jpg'))
        Image.new('L', (100, 100)).save(os.path.join(DATA_DIR, 'segmentation_task/test/masks', f's_{i}.png'))
# --- End: Fallback ---

SEG_TRAIN_IMAGES = os.path.join(DATA_DIR, 'segmentation_task/train/images')
SEG_TRAIN_MASKS = os.path.join(DATA_DIR, 'segmentation_task/train/masks')
SEG_TEST_IMAGES = os.path.join(DATA_DIR, 'segmentation_task/test/images')
SEG_TEST_MASKS = os.path.join(DATA_DIR, 'segmentation_task/test/masks')
CLF_TRAIN_DIR = os.path.join(DATA_DIR, 'classification_task/train')
CLF_TEST_DIR = os.path.join(DATA_DIR, 'classification_task/test')

IMAGE_SIZE = 256
# *** PRIMARY FIX: Drastically reduce Batch Size to prevent GPU Memory Error ***
# Original value was 16, then 8. A value of 4 is much safer for most GPUs.
BATCH_SIZE = 8
LEARNING_RATE = 1e-4
EPOCHS_JOINT = 40
EPOCHS_SEG_ONLY = 40
EPOCHS_CLF_ONLY = 40
NUM_CLASSES = 4
VALIDATION_SPLIT = 0.2

# --- Reproducibility ---
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    tf.random.set_seed(seed)
    os.environ['TF_DETERMINISTIC_OPS'] = '1'

set_seed(42)
gpus = tf.config.experimental.list_physical_devices('GPU')
if gpus:
    try:
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"Using GPU: {gpus[0].name}")
    except RuntimeError as e:
        print(e)
else:
    print("Using CPU")

# --- Data Loading and Preprocessing Functions (TensorFlow) ---
def parse_image_mask(img_path, mask_path):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, [IMAGE_SIZE, IMAGE_SIZE])
    img = (img / 127.5) - 1.0
    mask = tf.io.read_file(mask_path)
    mask = tf.image.decode_png(mask, channels=1)
    mask = tf.image.resize(mask, [IMAGE_SIZE, IMAGE_SIZE])
    mask = mask / 255.0
    return img, mask

def parse_image_label(img_path, label):
    img = tf.io.read_file(img_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, [IMAGE_SIZE, IMAGE_SIZE])
    img = (img / 127.5) - 1.0
    return img, label

def create_seg_dataset(img_dir, mask_dir, batch_size):
    img_paths = sorted(glob.glob(os.path.join(img_dir, '*.jpg')))
    mask_paths = [p.replace(img_dir, mask_dir).replace('.jpg', '.png') for p in img_paths]
    if not img_paths: return None, []
    dataset = tf.data.Dataset.from_tensor_slices((img_paths, mask_paths))
    dataset = dataset.map(parse_image_mask, num_parallel_calls=tf.data.AUTOTUNE)
    dataset = dataset.batch(batch_size).prefetch(buffer_size=tf.data.AUTOTUNE)
    return dataset, img_paths

def create_clf_dataset(data_dir, batch_size):
    classes = sorted([d for d in os.listdir(data_dir) if os.path.isdir(os.path.join(data_dir, d))])
    class_to_idx = {name: i for i, name in enumerate(classes)}
    img_paths, labels = [], []
    for cls_name in classes:
        cls_dir = os.path.join(data_dir, cls_name)
        for img_name in os.listdir(cls_dir):
            img_paths.append(os.path.join(cls_dir, img_name))
            labels.append(class_to_idx[cls_name])
    if not img_paths: return None, [], []
    path_label_pairs = list(zip(img_paths, labels))
    return path_label_pairs, classes

def get_split_datasets(path_label_pairs, batch_size, is_segmentation=False):
    if not path_label_pairs:
        return tf.data.Dataset.from_tensor_slices(([], [])).batch(batch_size), \
               tf.data.Dataset.from_tensor_slices(([], [])).batch(batch_size)
    random.shuffle(path_label_pairs)
    val_size = int(len(path_label_pairs) * VALIDATION_SPLIT)
    train_pairs = path_label_pairs[val_size:]
    val_pairs = path_label_pairs[:val_size]
    def create_pipeline(pairs):
        if not pairs:
            return tf.data.Dataset.from_tensor_slices(([], [])).batch(batch_size)
        paths, items = zip(*pairs)
        dataset = tf.data.Dataset.from_tensor_slices((list(paths), list(items)))
        if is_segmentation:
            dataset = dataset.map(parse_image_mask, num_parallel_calls=tf.data.AUTOTUNE)
        else:
            dataset = dataset.map(parse_image_label, num_parallel_calls=tf.data.AUTOTUNE)
        return dataset.batch(batch_size).prefetch(buffer_size=tf.data.AUTOTUNE)
    train_ds = create_pipeline(train_pairs)
    val_ds = create_pipeline(val_pairs)
    return train_ds, val_ds

# Data Loading
seg_img_paths = sorted(glob.glob(os.path.join(SEG_TRAIN_IMAGES, '*.jpg')))
seg_path_pairs = [(p, p.replace(SEG_TRAIN_IMAGES, SEG_TRAIN_MASKS).replace('.jpg', '.png')) for p in seg_img_paths]
seg_train_loader, seg_val_loader = get_split_datasets(seg_path_pairs, BATCH_SIZE, is_segmentation=True)
seg_test_loader, _ = create_seg_dataset(SEG_TEST_IMAGES, SEG_TEST_MASKS, BATCH_SIZE)
clf_path_label_pairs, clf_classes = create_clf_dataset(CLF_TRAIN_DIR, BATCH_SIZE)
clf_train_loader, clf_val_loader = get_split_datasets(clf_path_label_pairs, BATCH_SIZE)
clf_test_path_label_pairs, _ = create_clf_dataset(CLF_TEST_DIR, BATCH_SIZE)
if clf_test_path_label_pairs:
    test_paths, test_labels = zip(*clf_test_path_label_pairs)
    clf_test_loader = tf.data.Dataset.from_tensor_slices((list(test_paths), list(test_labels)))
    clf_test_loader = clf_test_loader.map(parse_image_label).batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
else:
    clf_test_loader = tf.data.Dataset.from_tensor_slices(([], [])).batch(BATCH_SIZE)


# --- Model Definition (Keras Functional API) ---
def DoubleConv(input_tensor, n_filters):
    x = layers.Conv2D(filters=n_filters, kernel_size=(3, 3), padding='same', use_bias=False)(input_tensor)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.Conv2D(filters=n_filters, kernel_size=(3, 3), padding='same', use_bias=False)(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    return x

def UNetWithClassifier(input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3), n_classes_clf=NUM_CLASSES):
    inputs = keras.Input(shape=input_shape)
    
    # *** SECONDARY FIX: Further reduce model complexity to guarantee it runs ***
    # This is a much lighter version of U-Net.
    f1, f2, f3, f4, f5 = 16, 32, 64, 128, 256

    # Encoder
    c1 = DoubleConv(inputs, f1)
    p1 = layers.MaxPooling2D((2, 2))(c1)
    c2 = DoubleConv(p1, f2)
    p2 = layers.MaxPooling2D((2, 2))(c2)
    c3 = DoubleConv(p2, f3)
    p3 = layers.MaxPooling2D((2, 2))(c3)
    c4 = DoubleConv(p3, f4)
    p4 = layers.MaxPooling2D((2, 2))(c4)
    c5 = DoubleConv(p4, f5)
    
    # Classifier Head
    clf_pool = layers.GlobalAveragePooling2D()(c5)
    clf_out = layers.Dense(n_classes_clf, name='clf_output')(clf_pool)
    
    # Decoder
    u6 = layers.Conv2DTranspose(f4, (2, 2), strides=(2, 2), padding='same')(c5)
    u6 = layers.concatenate([u6, c4])
    c6 = DoubleConv(u6, f4)
    u7 = layers.Conv2DTranspose(f3, (2, 2), strides=(2, 2), padding='same')(c6)
    u7 = layers.concatenate([u7, c3])
    c7 = DoubleConv(u7, f3)
    u8 = layers.Conv2DTranspose(f2, (2, 2), strides=(2, 2), padding='same')(c7)
    u8 = layers.concatenate([u8, c2])
    c8 = DoubleConv(u8, f2)
    u9 = layers.Conv2DTranspose(f1, (2, 2), strides=(2, 2), padding='same')(c8)
    u9 = layers.concatenate([u9, c1])
    c9 = DoubleConv(u9, f1)
    
    seg_out = layers.Conv2D(1, (1, 1), activation=None, name='seg_output')(c9)

    model = Model(inputs=[inputs], outputs=[seg_out, clf_out])
    return model

# --- Metrics and Visualization (Unchanged) ---
def iou_score(outputs, labels):
    outputs = tf.cast(tf.sigmoid(outputs) > 0.5, tf.float32)
    labels = tf.cast(labels > 0.5, tf.float32)
    intersection = tf.reduce_sum(outputs * labels, axis=[1, 2, 3])
    union = tf.reduce_sum(outputs + labels, axis=[1, 2, 3]) - intersection
    iou = (intersection + 1e-6) / (union + 1e-6)
    return tf.reduce_mean(iou)

def plot_loss_curves(train_losses, val_losses, title):
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Training Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

def display_prediction(model, image_path, true_mask_path=None, true_label=None, model_name="Model"):
    image = Image.open(image_path).convert("RGB")
    display_image = image.resize((IMAGE_SIZE, IMAGE_SIZE))
    display_tensor = np.array(display_image) / 255.0
    processed_tensor = (tf.convert_to_tensor(np.array(display_image), dtype=tf.float32) / 127.5) - 1.0
    processed_tensor = tf.expand_dims(processed_tensor, 0)
    pred_mask_tensor, pred_class_logits = model(processed_tensor, training=False)
    pred_mask_binary = (tf.sigmoid(pred_mask_tensor[0]) > 0.5).numpy().astype(np.uint8)
    pred_class_idx = tf.argmax(pred_class_logits[0]).numpy()
    pred_class_name = clf_classes[pred_class_idx]
    num_plots = 3 if true_mask_path else 2
    plt.figure(figsize=(6 * num_plots, 6))
    plt.subplot(1, num_plots, 1)
    plt.imshow(display_tensor)
    true_label_str = f"True: {clf_classes[true_label]}" if true_label is not None else ""
    plt.title(f"Input Image ({model_name})\nPredicted: {pred_class_name}\n{true_label_str}")
    plt.axis('off')
    plt.subplot(1, num_plots, 2)
    plt.imshow(pred_mask_binary[:, :, 0], cmap='gray')
    plt.title("Predicted Mask")
    plt.axis('off')
    if true_mask_path:
        true_mask = Image.open(true_mask_path).convert("L").resize((IMAGE_SIZE, IMAGE_SIZE))
        plt.subplot(1, num_plots, 3)
        plt.imshow(true_mask, cmap='gray')
        plt.title("Ground Truth Mask")
        plt.axis('off')
    plt.show()
    
# --- Training Strategies (Unchanged Logic) ---
seg_criterion = losses.BinaryCrossentropy(from_logits=True)
clf_criterion = losses.SparseCategoricalCrossentropy(from_logits=True)

def train_joint(model, seg_train_loader, seg_val_loader, clf_train_loader, clf_val_loader, optimizer, num_epochs):
    print("\n--- Starting Joint Training ---")
    history = {'train_seg_loss': [], 'val_seg_loss': [], 'train_clf_loss': [], 'val_clf_loss': []}
    
    @tf.function
    def train_step(seg_batch, clf_batch):
        seg_inputs, seg_masks = seg_batch
        clf_inputs, clf_labels = clf_batch
        with tf.GradientTape() as tape:
            seg_outputs, _ = model(seg_inputs, training=True)
            _, clf_outputs = model(clf_inputs, training=True)
            seg_loss = seg_criterion(seg_masks, seg_outputs)
            clf_loss = clf_criterion(clf_labels, clf_outputs)
            total_loss = seg_loss + clf_loss
        grads = tape.gradient(total_loss, model.trainable_variables)
        optimizer.apply_gradients(zip(grads, model.trainable_variables))
        return seg_loss, clf_loss

    for epoch in range(num_epochs):
        running_seg_loss, running_clf_loss, num_batches = 0.0, 0.0, 0
        train_loader = tf.data.Dataset.zip((seg_train_loader, clf_train_loader))
        progress_bar = tqdm(train_loader, desc=f"Joint Epoch {epoch+1}/{num_epochs}", leave=True)
        for i, (seg_batch, clf_batch) in enumerate(progress_bar):
            num_batches = i + 1
            seg_loss, clf_loss = train_step(seg_batch, clf_batch)
            running_seg_loss += seg_loss; running_clf_loss += clf_loss
            progress_bar.set_postfix(seg_loss=f"{running_seg_loss / num_batches:.4f}", clf_loss=f"{running_clf_loss / num_batches:.4f}")
        if num_batches == 0: continue
        
        val_seg_loss, val_clf_loss = 0.0, 0.0
        len_seg_val = 0
        for _ in seg_val_loader: len_seg_val += 1
        for seg_inputs, seg_masks in seg_val_loader:
            seg_outputs, _ = model(seg_inputs, training=False); val_seg_loss += seg_criterion(seg_masks, seg_outputs)
        
        len_clf_val = 0
        for _ in clf_val_loader: len_clf_val += 1
        for clf_inputs, clf_labels in clf_val_loader:
            _, clf_outputs = model(clf_inputs, training=False); val_clf_loss += clf_criterion(clf_labels, clf_outputs)

        avg_train_seg, avg_train_clf = running_seg_loss / num_batches, running_clf_loss / num_batches
        avg_val_seg = val_seg_loss / len_seg_val if len_seg_val > 0 else 0
        avg_val_clf = val_clf_loss / len_clf_val if len_clf_val > 0 else 0

        history['train_seg_loss'].append(avg_train_seg); history['val_seg_loss'].append(avg_val_seg)
        history['train_clf_loss'].append(avg_train_clf); history['val_clf_loss'].append(avg_val_clf)
        print(f"Epoch {epoch+1} Summary: Train Seg: {avg_train_seg:.4f}, Val Seg: {avg_val_seg:.4f}, Train Clf: {avg_train_clf:.4f}, Val Clf: {avg_val_clf:.4f}")

        if (epoch + 1) % 10 == 0 and clf_path_label_pairs:
            print(f"\n--- Predictions after Epoch {epoch+1} ---")
            img_path, true_label = random.choice(clf_path_label_pairs)
            display_prediction(model, img_path, true_label=true_label, model_name=f"Joint (Epoch {epoch+1})")
            
    plot_loss_curves(history['train_seg_loss'], history['val_seg_loss'], "Joint Training - Segmentation Loss")
    plot_loss_curves(history['train_clf_loss'], history['val_clf_loss'], "Joint Training - Classification Loss")
    print("Finished Joint Training.")
    return model

def train_separate(model, seg_train_loader, seg_val_loader, clf_train_loader, clf_val_loader, optimizer, epochs_seg, epochs_clf):
    print("\n--- Starting Separate Training ---")
    
    print("\n--- Stage 1: Training Segmentation Only ---")
    seg_history = {'train_loss': [], 'val_loss': []}
    @tf.function
    def seg_train_step(inputs, masks):
        with tf.GradientTape() as tape:
            seg_outputs, _ = model(inputs, training=True); loss = seg_criterion(masks, seg_outputs)
        grads = tape.gradient(loss, model.trainable_variables); optimizer.apply_gradients(zip(grads, model.trainable_variables))
        return loss
        
    for epoch in range(epochs_seg):
        running_loss, num_batches = 0.0, 0
        for i, (inputs, masks) in enumerate(tqdm(seg_train_loader, desc=f"Seg-Only Epoch {epoch+1}/{epochs_seg}")):
            num_batches = i + 1; running_loss += seg_train_step(inputs, masks)
        if num_batches == 0: continue
        val_loss, len_seg_val = 0.0, 0
        for _ in seg_val_loader: len_seg_val += 1
        for inputs, masks in seg_val_loader:
            seg_outputs, _ = model(inputs, training=False); val_loss += seg_criterion(masks, seg_outputs)
        avg_train, avg_val = (running_loss/num_batches), (val_loss/len_seg_val if len_seg_val > 0 else 0)
        seg_history['train_loss'].append(avg_train); seg_history['val_loss'].append(avg_val)
        print(f"Epoch {epoch+1} Summary: Train Loss: {avg_train:.4f}, Val Loss: {avg_val:.4f}")
        if (epoch + 1) % 10 == 0 and seg_path_pairs:
             img_path, mask_path = random.choice(seg_path_pairs)
             display_prediction(model, img_path, true_mask_path=mask_path, model_name=f"Seg-Only (Epoch {epoch+1})")
    
    plot_loss_curves(seg_history['train_loss'], seg_history['val_loss'], "Separate Training - Segmentation Loss")

    print("\n--- Freezing All But Classifier ---")
    for layer in model.layers:
        if 'clf' not in layer.name: layer.trainable = False
    
    print("\n--- Stage 2: Training Classifier Only ---")
    clf_history = {'train_loss': [], 'val_loss': []}
    @tf.function
    def clf_train_step(inputs, labels):
        with tf.GradientTape() as tape:
            _, clf_outputs = model(inputs, training=True); loss = clf_criterion(labels, clf_outputs)
        grads = tape.gradient(loss, model.trainable_variables); optimizer.apply_gradients(zip(grads, model.trainable_variables))
        return loss
        
    for epoch in range(epochs_clf):
        running_loss, num_batches = 0.0, 0
        for i, (inputs, labels) in enumerate(tqdm(clf_train_loader, desc=f"Clf-Only Epoch {epoch+1}/{epochs_clf}")):
            num_batches = i + 1; running_loss += clf_train_step(inputs, labels)
        if num_batches == 0: continue
        val_loss, len_clf_val = 0.0, 0
        for _ in clf_val_loader: len_clf_val += 1
        for inputs, labels in clf_val_loader:
            _, clf_outputs = model(inputs, training=False); val_loss += clf_criterion(labels, clf_outputs)
        avg_train, avg_val = running_loss/num_batches, (val_loss/len_clf_val if len_clf_val > 0 else 0)
        clf_history['train_loss'].append(avg_train); clf_history['val_loss'].append(avg_val)
        print(f"Epoch {epoch+1} Summary: Train Loss: {avg_train:.4f}, Val Loss: {avg_val:.4f}")
        if (epoch + 1) % 10 == 0 and clf_path_label_pairs:
            img_path, true_label = random.choice(clf_path_label_pairs)
            display_prediction(model, img_path, true_label=true_label, model_name=f"Clf-Only (Epoch {epoch+1})")

    plot_loss_curves(clf_history['train_loss'], clf_history['val_loss'], "Separate Training - Classification Loss")
    for layer in model.layers: layer.trainable = True
    print("Finished Separate Training.")
    return model

def evaluate_model(model, seg_loader, clf_loader, model_name="Model"):
    print(f"\n--- Evaluating {model_name} ---")
    total_iou, num_batches = 0.0, 0
    if seg_loader:
        for _ in seg_loader: num_batches += 1
        for inputs, masks in tqdm(seg_loader, desc="Evaluating Segmentation"):
            outputs, _ = model(inputs, training=False); total_iou += iou_score(outputs, masks)
        avg_iou = total_iou / num_batches if num_batches > 0 else 0
    else: avg_iou = 0
    all_preds, all_labels = [], []
    for inputs, labels in tqdm(clf_loader, desc="Evaluating Classification"):
        _, outputs = model(inputs, training=False); all_preds.extend(tf.argmax(outputs, axis=1).numpy()); all_labels.extend(labels.numpy())
    print(f"\n--- Performance for {model_name} ---")
    print(f"Average IoU on Test Set: {avg_iou:.4f}")
    if all_labels:
        print("\nClassification Report:"); print(classification_report(all_labels, all_preds, target_names=clf_classes, zero_division=0))
        print("\nConfusion Matrix:"); cm = confusion_matrix(all_labels, all_preds)
        plt.figure(figsize=(8, 6)); sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=clf_classes, yticklabels=clf_classes)
        plt.xlabel('Predicted'); plt.ylabel('True'); plt.title(f'Confusion Matrix for {model_name}'); plt.show()

# --- Main Execution ---
print("\n" + "="*50 + "\nINITIALIZING JOINT TRAINING\n" + "="*50)
K.clear_session(); gc.collect()
model_joint = UNetWithClassifier(n_classes_clf=NUM_CLASSES)
optimizer_joint = optimizers.Adam(learning_rate=LEARNING_RATE)
model_joint = train_joint(model_joint, seg_train_loader, seg_val_loader, clf_train_loader, clf_val_loader, optimizer_joint, num_epochs=EPOCHS_JOINT)
evaluate_model(model_joint, seg_test_loader, clf_test_loader, model_name="Jointly Trained Model")

print("\n" + "="*50 + "\nINITIALIZING SEPARATE TRAINING\n" + "="*50)
K.clear_session(); gc.collect()
model_separate = UNetWithClassifier(n_classes_clf=NUM_CLASSES)
optimizer_separate = optimizers.Adam(learning_rate=LEARNING_RATE)
model_separate = train_separate(model_separate, seg_train_loader, seg_val_loader, clf_train_loader, clf_val_loader, optimizer_separate, epochs_seg=EPOCHS_SEG_ONLY, epochs_clf=EPOCHS_CLF_ONLY)
evaluate_model(model_separate, seg_test_loader, clf_test_loader, model_name="Separately Trained Model")

print("\n" + "="*50 + "\nDISPLAYING FINAL SAMPLE PREDICTIONS ON TEST DATA\n" + "="*50)
if clf_test_path_label_pairs:
    random_image_path, true_label = random.choice(clf_test_path_label_pairs)
    print("\n--- Prediction from Jointly Trained Model ---")
    display_prediction(model_joint, random_image_path, true_label=true_label, model_name="Final Joint Model")
    print("\n--- Prediction from Separately Trained Model ---")
    display_prediction(model_separate, random_image_path, true_label=true_label, model_name="Final Separate Model")
else:
    print("No images in the test set to display.")
```
